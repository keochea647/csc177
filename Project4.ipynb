# ANN Section (Part 3)

```python
from keras import Sequential
from keras.layers import Dense
import numpy as np
from sklearn import preprocessing

filename_read2 = os.path.join(path,"Admission_Predict_Ver1.1_small_data_set_for_Linear_Regression.csv")
classDF = pd.read_csv(filename_read2, na_values=['NA','?'], encoding="ISO-8859-1")
classDF.drop(['Serial No.'],inplace=True,axis=1)
print('Number of instances = %d' % (classDF.shape[0]))
print('Number of attributes = %d' % (classDF.shape[1]))
print('Number of missing values:')
for col in classDF.columns:
    print('\t%s: %d' % (col,classDF[col].isna().sum()))
classDF.head()
```

```python
lab1 = "Less than 60%"
lab2 = "More than or equal to 60%"
classDF.loc[classDF['Chance of Admit '].between(0,0.6, 'left'),'Admission Chance'] = lab1;
classDF.loc[classDF['Chance of Admit '].between(0.6,1,'both'),'Admission Chance'] = lab2;
classDF.drop(['Chance of Admit '],inplace=True,axis=1)
classDF
```

```python
# Encode text values to indexes(i.e. [1],[2],[3] for red,green,blue).
def encode_text_index(df, name):
    le = preprocessing.LabelEncoder()
    df[name] = le.fit_transform(df[name])
    return le.classes_
```

```python
import collections
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Convert a Pandas dataframe to the x,y inputs that TensorFlow needs
def to_xy(df, target):
    result = []
    for x in df.columns:
        if x != target:
            result.append(x)
    # find out the type of the target column. 
    target_type = df[target].dtypes
    target_type = target_type[0] if isinstance(target_type, collections.Sequence) else target_type
    # Encode to int for classification, float otherwise. TensorFlow likes 32 bits.
    if target_type in (np.int64, np.int32):
        # Classification
        dummies = pd.get_dummies(df[target])
        return df[result].values.astype(np.float32), dummies.values.astype(np.float32)
    else:
        # Regression
        return df[result].values.astype(np.float32), df[target].values.astype(np.float32)

#Function to normalize columns
def normalize_numeric_minmax(df, name):
        df[name] = ((df[name] - df[name].min()) / (df[name].max() - df[name].min())).astype(np.float32)

# normalizing inputs
for cols in classDF.columns:
  if(cols != 'Research' and cols != 'Admission Chance'):
    normalize_numeric_minmax(classDF,cols);


# not really sure if we should split into test and train data
#Classes = encode_text_index(classDF, 'Admission Chance')
#X,Y = to_xy(classDF,'Admission Chance')

X = classDF.drop(['Admission Chance'], axis=1)
Y = classDF['Admission Chance']
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=69)

# handling imbalanced classes
sm = SMOTE(random_state=27)
x_train, y_train = sm.fit_resample(x_train, y_train)

trainDF = x_train
trainDF['Admission Chance'] = y_train
testDF = x_test
testDF['Admission Chance'] = y_test

Classes = encode_text_index(trainDF, 'Admission Chance')
encode_text_index(testDF, 'Admission Chance')

x_train, y_train = to_xy(trainDF, 'Admission Chance')
x_test, y_test = to_xy(testDF, 'Admission Chance')

print(x_train.shape)
print(y_train.shape)
```

```python
Classes
```

```python
model = Sequential()
model.add(Dense(12, input_dim = x_train.shape[1], activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam')
model.fit(x_train,y_train,verbose=2, epochs=100)
```

```python
pred = model.predict(x_test)
print(pred[0])
```

```python
pred = np.argmax(pred, axis=1)
```

```python
true = np.argmax(y_test, axis=1)
```

```python
# Plot a confusion matrix.
# cm is the confusion matrix, names are the names of the classes.
def plot_confusion_matrix(cm, names, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(names))
    plt.xticks(tick_marks, names, rotation=45)
    plt.yticks(tick_marks, names)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

predClassY = clf.predict(x2_test)
print('Accuracy on test data is %.2f' % (accuracy_score(y2_test, predClassY)))
```

```python
confusion = confusion_matrix(y2_test, predClassY)
print(confusion)

plot_confusion_matrix(confusion, [lab1,lab2], title='Confusion matrix', cmap=plt.cm.Blues)

print('Accuracy on test data is %.2f' % (accuracy_score(y2_test, predClassY)))
print('F1 score on test data is %.2f' % (f1_score(y2_test, predClassY, average='weighted')))
print('Precision Score on test data is %.2f' % (precision_score(y2_test, predClassY, average='weighted')))
print('Recall score on test data is %.2f' % (recall_score(y2_test, predClassY, average='weighted')))
print( classification_report(y2_test,predClassY))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score,recall_score,classification_report

C = [0.01, 0.1, 0.2, 0.5, 0.8, 1, 5, 10, 20, 50]

LRtestAcc = []
LRtrainAcc = []

for param in C:
    clf = LogisticRegression(C=param)
    clf.fit(X2,Y2)
    log_reg_pred = clf.predict(x2_test)
    log_reg_pred_train = clf.predict(X2)
    print(log_reg_pred)
    LRtestAcc.append(accuracy_score(y2_test, log_reg_pred))
    LRtrainAcc.append(accuracy_score(Y2,log_reg_pred_train))

plt.plot(C, LRtestAcc,'bv--',C,LRtrainAcc,'ro--')
plt.legend(['Test Accuracy','Train Accuracy'])
plt.xlabel('C')
plt.xscale('log')
plt.ylabel('Accuracy') 
```

```python
from sklearn.neighbors import KNeighborsClassifier

numNeighbors = [1, 5, 10]
testAcc = []
trainAcc = []

for k in numNeighbors:
    clf = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p=2)
    clf.fit(X2, Y2)
    knn_pred = clf.predict(x2_test)
    knn_pred_train = clf.predict(X2)
    print(knn_pred)
    testAcc.append(accuracy_score(y2_test, knn_pred))
    trainAcc.append(accuracy_score(Y2,knn_pred_train))

plt.plot(numNeighbors, testAcc,'bv--',numNeighbors, trainAcc, 'ro--')
plt.legend(['Test Accuracy','Train Accuacy'])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
```
